```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
from sklearn import linear_model
```

```python
data=pd.read_csv('pizza.csv')
```

```python
data
```

```python
 sns.lmplot(x='diameter',y='price',data=data)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(data[['diameter']],data['price'])
```

```python
reg.predict([[20]])
```

```python
reg.predict([[50]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
